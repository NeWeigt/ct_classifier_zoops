```python
import os
import torch

# Path to your project root
project_root = "/home/Nele/code/ct_classifier_zoops"

# Paths to the "best epoch" checkpoints
paths = {
    "ResNet-18 (model.py)": os.path.join(project_root, "best_epochs", "first_training_resnet18.pt"),
    "ResNet-50 (bigmodel.py)": os.path.join(project_root, "best_epochs", "first_training_resnet50.pt"),
    "ResNet-50 (more stations)": os.path.join(project_root, "best_epochs", "morestations_resnet50.pt"),
}

# Load validation accuracies
results = {}
for name, path in paths.items():
    ckpt = torch.load(path, map_location="cpu")
    results[name] = ckpt["oa_val"]

# Print all results
print("Validation accuracy (oa_val):")
for name, acc in results.items():
    print(f"  {name:<30} : {acc:.4f}")

# Find best model
best_model = max(results, key=results.get)
print(f"Best model on validation: {best_model} ({results[best_model]:.4f})")
```
