```python
## which model is better? RestNet18 or ResNet20?
import os
import torch

# Path to your project root
project_root = "/home/Nele/code/ct_classifier_zoops"

# Paths to the "best epoch" checkpoints you already saved
path_resnet18 = os.path.join(project_root, "best_epochs", "first_training_resnet18.pt")
path_resnet50 = os.path.join(project_root, "best_epochs", "first_training_resnet50.pt")

# Load checkpoints on CPU (we only need the metrics)
ckpt18 = torch.load(path_resnet18, map_location="cpu")
ckpt50 = torch.load(path_resnet50, map_location="cpu")

acc18 = ckpt18["oa_val"]   # validation accuracy for ResNet-18
acc50 = ckpt50["oa_val"]   # validation accuracy for ResNet-50

print(f"ResNet-18 (model.py)  oa_val: {acc18:.4f}")
print(f"ResNet-50 (bigmodel.py) oa_val: {acc50:.4f}")

if acc18 > acc50:
    print("ResNet-18 performed better on validation.")
elif acc50 > acc18:
    print("ResNet-50 performed better on validation.")
else:
    print("Both models performed equally well on validation.")
```
