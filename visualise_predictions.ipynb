```python
import json
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import confusion_matrix
from pathlib import Path
```

```python
fp = "/home/Nele/code/ct_classifier_zoops/val_predictions_epoch_.json"
```

```python
# load data from JSON file
for i in [1,21,41,61,81, ]:
    fp = f"/home/Nele/code/ct_classifier_zoops/val_predictions_epoch_{i}.json"
    with open(fp, 'r') as f:
        data = json.load(f)
        predictions = data['predictions']
        ground_truths = data['labels']

    # convert lists to numpy arrays for easier manipulation
    predictions = np.array(predictions)
    ground_truths = np.array(ground_truths)

    # plot confusion matrix

    cm = confusion_matrix(ground_truths, predictions)
    plt.figure(figsize=(8, 6))
    plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
    plt.title(f'Confusion Matrix ep:{i}')
    plt.colorbar()
    tick_marks = np.arange(len(set(ground_truths)))
    plt.xticks(tick_marks, tick_marks)
    plt.yticks(tick_marks, tick_marks)
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.tight_layout()
    plt.show()
    # plt.savefig('confusion_matrix.png')
```
